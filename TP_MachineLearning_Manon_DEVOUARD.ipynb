**Partie 2 :**

*2.1 Gestion des valeurs manquantes*

```python
missing_values = df.isnull().sum()
for column in df.columns:
    if df[column].dtype == 'object':
        df[column].fillna(df[column].mode()[0], inplace=True)
    else:
        df[column].fillna(df[column].median(), inplace=True)
```

```python
df
```

Pour gérer les valeurs manquantes, j'ai choisi d'éstimer ces valeurs en faisant une imputation par la moyenne puisqu'il y a déjà des valeurs numériques dans les colonnes. Pour ça, on prend toutes les valeurs de la colonne et on prend la moyenne.


*2.2 Encodage des catégories*

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
```

```python
data = df["tartar"]
data2 = df["gender"]
data3 = df["oral"]
print (data)
label_encoder = LabelEncoder()
integer_encoded = label_encoder.fit_transform(data)

onehot_encoder = OneHotEncoder()

encoded_tartar = onehot_encoder.fit_transform(df[['tartar']]).toarray()
encoded_gender = onehot_encoder.fit_transform(df[['gender']]).toarray()
encoded_oral = onehot_encoder.fit_transform(df[['oral']]).toarray()

encoded_tartar_df = pd.DataFrame(encoded_tartar, columns=[f"tartar_{i}" for i in range(encoded_tartar.shape[1])])
encoded_gender_df = pd.DataFrame(encoded_gender, columns=[f"gender_{i}" for i in range(encoded_gender.shape[1])])
encoded_oral_df = pd.DataFrame(encoded_oral, columns=[f"oral_{i}" for i in range(encoded_oral.shape[1])])

df = pd.concat([df.drop(columns=['tartar', 'gender', 'oral']), encoded_tartar_df, encoded_gender_df, encoded_oral_df], axis=1)
```

**Partie 3**

*3.1 Création de jeu de test et d'entraînemente*

```python
X = df.drop('smoking', axis=1)
y=df ['smoking']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

Nous avons utilisé 80% des données pour l'entraînement et 20% pour le test.

Il est important de diviser les données pour créer un ensemble de tests indépendants pour évaluer la performance du modèle. Cela aide à vérifier si le modèle généralise bien sur des données qu'il n'a pas vues pendant l'entraînement.

*3.2 Entraînement d'un modèle simple : LogisticRegression*

```python
X.count()
y.count()
```

```python
y_pred = clf.predict(X_test)
```
